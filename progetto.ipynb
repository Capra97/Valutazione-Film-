```python
def correggi_segmento(testo):
    if ', ' in testo:
        before, after = testo.rsplit(', ', 1)
        return f"{after} {before}"
    return testo
def correggi_titolo(titolo):
    if pd.isna(titolo):
        return titolo
    titolo = titolo.strip('"')  # Rimuove virgolette manuali
    # Se contiene una parte tra parentesi
    match = re.match(r'^(.*?)(\s*\(.*\))$', titolo)
    if match:
        parte1 = match.group(1)
        parte2 = match.group(2)
        nuova_parte1 = correggi_segmento(parte1)
        contenuto = parte2[1:-1].strip()
        contenuto = contenuto.replace("(", "").replace(")", "")
        nuova_parte2 = correggi_segmento(contenuto)
        return f'{nuova_parte1} ({nuova_parte2})'
    else:
        nuova_parte1 = correggi_segmento(titolo)
        return nuova_parte1
# Applica la funzione
df['Title'] = df['Title'].apply(correggi_titolo)
# Salva il CSV correttamente (senza virgolette doppie)
df.to_csv("movies_corretto.csv", index=False, quoting=0)
```

```python
def capitalizza_generi(generi):
    if pd.isna(generi):
        return generi
    return '|'.join(g.strip().capitalize() for g in generi.split('|'))
df['Genres'] = df['Genres'].apply(capitalizza_generi)
df.to_csv("movies_corretto.csv", index=False, quoting=0)
```
